```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/sagnikghoshcr7/Bank-Customer-Churn-Prediction/master/Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
## Dropping irrelevant columns
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df["Geography"].value_counts()
```

```python
df.info()
```

```python
df["Gender"].value_counts()
```

```python
df=pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)
```

```python
df.head()
```

```python
# Dep Var="Exited"
```

```python
X=df.drop(["Exited"],axis=1)
```

```python
y=df["Exited"]
```

```python
import tensorflow
from tensorflow import keras
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Sequential() is a class
```

```python
model3=Sequential()
```

```python

model3.add(Dense(11,activation="sigmoid",input_dim=11)) #Input Layer + Hidden with 11 Nodes

model3.add(Dense(11,activation="sigmoid"))

model3.add(Dense(1,activation="sigmoid")) #Output Layer
```

```python
len(X.columns)
```

```python
model3.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scale=scaler.fit_transform(X_train)
X_test_scale=scaler.transform(X_test)
```

```python
model3.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model3.fit(X_train_scale,y_train,epochs=100,verbose=1)
```

```python
y_pred=model3.predict(X_test_scale)
```

```python
y_pred
```

```python
y_pred=y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
